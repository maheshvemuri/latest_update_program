**advantages**
- using one hot encoder we will get new columns
- the new column values at a time only one will be **True** ,another will be **False**
- which means new columns are ** independent each other**
- independent means **no relation**
- means no **multi collineraity**
- multicollineraity is an important property in machine learning 
- **machine learning models expert the input columns should be independent each other**
- so one hot encoder give the no multicollineraity

** disadvantage**

- the disadvantage is it will increase the many columns
- in the above example the total columns be come 25510
- this is happend because of case id
- in real time we will drop the id columns
- eventhought if we drop still we are getting 30 columns
- more columns means more time to process
- we required more infrastaructur (cpu- gpu -tpu)
- this is called as **curse of dimensionality**
